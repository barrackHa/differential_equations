```python
%matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import models as m
```

```python
## Pitchfork
time = np.arange(0, 9, 0.001)
r0 = -3
x0 = 0.1
a = 25
epsilon = 1.5
sigma = 0.3
config = {
    'time': time,
    'y0': [x0, r0],
    'epsilon': epsilon,
    'sigma_noise': [sigma, 0],
    'a': a,
    'ttl': r"Pitchfork: $\dot{x}$ = rx - $ax^3$,    $\dot{r}$ = $\epsilon$" + "\n" 
}

fig, axs = plt.subplots(1,1)

for i in range(10):
    scp_model = m.SuperCriticalPitchfork()
    res, des = scp_model.simulate(config=config)
    fig, axs, t_star = scp_model.plot(fig=fig, axs=axs)

scp_model._plot_roots_of_pitchfork(axs[0])

plt.show()
```
